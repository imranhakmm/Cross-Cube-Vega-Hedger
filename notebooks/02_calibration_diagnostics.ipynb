# Calibration Diagnostics

Per-slice SABR calibration enforces the ATM vol exactly and fits rho and nu.

```python
from cxvega.calibration import calibrate_cube_per_slice
from cxvega.config import load_settings
from cxvega.simulator import simulate_cube_path

settings = load_settings('../configs/default.yaml')
path = simulate_cube_path(settings, n_steps=2)
result = calibrate_cube_per_slice(path.forwards, path.strikes, path.expiries, path.vols[0], path.beta)
float(abs(result.residuals).max())
```
